```python
d = {"HeartAttack":"Cardiac_Arrest", "Gender":"Sex","SmokingHabits":"Smoker","AlcoholConsumption":"Alcoholic"}
df.rename(columns= d , inplace=True )
```

```python

df['Alcoholic'].fillna('Never', inplace=True)

df.isna().sum()
```

```python
df.head(5)
```

```python
le = LabelEncoder()
df['Cardiac_Arrest'] = le.fit_transform(df['Cardiac_Arrest'])
df['Smoker'] = le.fit_transform(df['Smoker'])
df['Alcoholic'] = le.fit_transform(df['Alcoholic'])
df['Sex'] = le.fit_transform(df['Sex'])
df['Hypertension'] = le.fit_transform(df['Hypertension'])
df.head()
```

```python
df['Alcoholic'].unique()
```

```python
ss = StandardScaler()

df[['Age','Cholesterol']] = ss.fit_transform(df[['Age','Cholesterol']])

df.head(5)
```

```python
X = df[['Sex','Age','Cholesterol','Smoker','Hypertension','Alcoholic']]
Y = df['Cardiac_Arrest']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,train_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train,y_train)
```

```python
model_rf = RandomForestClassifier()
model_rf.fit(X_train,y_train)
```
